### 1. Write a program that swaps the values of variables a and b. You are not allowed to use the third variable. You are not allowed to perform any arithmetics on a and b

```python
a, b = 5, 10
print(f"a={a}\tb={b}")
b, a = a, b
print(f"a={a}\tb={b}")
```

### 2. Write a program that generates 5 random numbers in the range 10 to 50. Use a seed value of 6. Make a provision to change this seed value every time you execute the program by associate it with time of execution.

```python
import random
random.seed(6) # seed associated with value of 6
nums = [random.randint(10, 50) for i in range(5)]
print(nums)

random.seed() # seed associated with time of execution
nums = [random.randint(10, 50) for i in range(5)]
print(nums)

nums = [random.randint(10, 50) for i in range(5)]
print(nums)
```

### 3. Use trunc(). floor() and ceil() for numbers -2.8, -0.5, 0.2, 1.5 and 2.9 to understand the difference between these functions clearly

```python
import math
nums = [-2.8, -0.5, 0.2, 1.5, 2.9]
print("num\ttrunc\tfloor\tceil")
for num in nums:
    print(f"{num}\t{math.trunc(num)}\t{math.floor(num)}\t{math.ceil(num)}")
```

### 4. Assume a suitable value for Dinesh's basic salary. His dearness allowance is 40% of basic salary and house rent allowance is 20% of basic salary. Write a program to calculate his gross salary.

```python
basic_salary = 75000
dallowance = 0.4 * basic_salary
rallowance = 0.2 * basic_salary
gross_salary = basic_salary + dallowance + rallowance
print(f"Basic Salary : {basic_salary}")
print(f"Dearness Allowance : {dallowance}")
print(f"House Rent : {rallowance}")
print(f"Gross Salary : {gross_salary}")
```
