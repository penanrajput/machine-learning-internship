### 1) Any year is input through the keyboard. Write a program to determine whether the year is a leap year or not.

### 2) Three numbers are entered throgh keyboard. Write a program to print smallest number of them.

```python
nums = [int(num) for num in input().split()]
fno, sno, tno = nums[0], nums[1], nums[2]

result = min(fno, min(sno, tno))
print(f"Nums : {nums}\tSmallest : {result}")
```

### 4) Given the coordinates (x,y) of center of a circle and its radius. Write a program that will determine whether a point lies insode the circle, on the circle or outside the circle. (Use sqrt() and pow() functions)


```python
c = [int(num) for num in input().split()]
c_x, c_y, radius = c[0], c[1], c[2]

pt = [int(num) for num in input().split()]
p_x, p_y = pt[0], pt[1]

dist = (p_y-c_y) ** 2 + (p_x-c_x)**2
ans = dist < radius * radius

print(f"Circle {c}, point {pt}, is point inside circle ? {ans}")


# lies inside <
# on circle ==
# lies inside or on circle ==
# lies outside  >
# lies outside or on circle >=
```

### 6) Write a program that prints all unique combinations of 1, 2 and 3

```python
nums = [1, 2, 3]
for i in range(3):
    for j in range(3):
        for k in range(3):
            if all([i != j, j != k, i != k]):
                print(f"{nums[i]}{nums[j]}{nums[k]}", end=" ")
```

### 7) Write a program that receives an integer and determine whether it is a prime number or not

### 9) Write a program to count the number of alphabets and number of digits in the string 'Maharashtra-422001'

```python
str = "Maharashtra-422001"
alphas = digits = 1
for char in str:
    if char.isalpha():
        alphas += 1
    if char.isdigit():
        digits += 1
print(f"string\t\t\tAlphabets\tDigits")
print(f"{str}\t{alphas}\t\t{digits}")
```

### 10) Write a program to reverse the String

```python
str = "Maharashtra-422001"
rev = ""
for char in str:
    rev = char + rev

print(f"String\t\t\tReverse")
print(f"{str}\t{rev}")
```

### 11) Write a program to print all prime numbers from 1 to 500

```python
for num in range(1, 500):
    isPrime = True
    if (num % 2 == 0 and num != 2) or (num <= 1):
        isPrime = False
    else:
        for i in range(3, num//2+1, 2):
            if num % i == 0:
                isPrime = False
                break
    # print(f"{num}, is Prime ? {isPrime}")
    if isPrime == True:
        print(num, end=" ")
```

### 12) Ramanujan number is the smallest number that can be expressed as sum of two cubes in two different ways. Write a program to print all such numbers up to a reasonable limit.

ref: https://www.ndtv.com/education/national-mathematics-day-2019-hardy-ramanujan-number-2152767



```python
limit = 40
nums = [i for i in range(1, limit+1)]
cubes = [i**3 for i in range(1, limit+1)]
# print(nums)
# print(cubes)

ramanujan = dict()

for first in range(len(nums)):
    for second in range(len(nums)-1, first, -1):
        if first < second:
            cube = cubes[first] + cubes[second]
            if cube in ramanujan:
                ramanujan[cube] += 1
                # print(first, second, cube)
                print(cube, end=" ")
            else:
                ramanujan[cube] = 1

# # to print the values
# for (key, value) in ramanujan.items():
#     if value > 1:
#         print(key, value)
```
